```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df=pd.read_csv("test.csv")
df.head()
```

```python
cat_nan_features=[feature for feature in df.columns if df[feature].isnull().sum()>0 and df[feature].dtypes=='O']
for feature in cat_nan_features:
    print(feature,np.round(df[feature].isnull().mean(),4),'% missing values')
```

```python
def fill_cat_nan(data,nan_features):
    data=df.copy()
    data[nan_features]=data[nan_features].fillna("Missing")
    return data
df=fill_cat_nan(df,cat_nan_features)
df[cat_nan_features].isnull().sum()
```

```python
num_nan=[feature for feature in df.columns if df[feature].isnull().sum()>0 and df[feature].dtypes!='O']
for feature in num_nan:
    print(feature,np.round(df[feature].isnull().mean(),4),'% missing values')
```

```python
for feature in num_nan:
    median_val=df[feature].median()
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature]=df[feature].fillna(median_val)
df[num_nan].isnull().sum()
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
for feature in num_features:
    df[feature]=np.log(df[feature])
df[['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']].head()
```
