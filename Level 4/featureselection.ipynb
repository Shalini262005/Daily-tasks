```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
df=pd.read_csv("X_train.csv")
df.head()
```

```python
Y_train=df['SalePrice']
Y_train
```

```python
X_train=df.drop(['Id', 'SalePrice'],axis=1)
X_train
```

Selecting alpha value for Lasso 

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv=LassoCV(alphas=[0.001,0.005,0.01,0.05,0.1],cv=5,random_state=0)
lasso_cv.fit(X_train,Y_train)
print(f'Optimal alpha: {lasso_cv.alpha_}')
```

```python
lasso=SelectFromModel(Lasso(alpha=0.001,random_state=0))
lasso.fit(X_train,Y_train)
```

```python
lasso.get_support()
```

```python
sel_features=X_train.columns[(lasso.get_support())]
print(f'Total features: {X_train.shape[1]}')
print(f'Selected features: {len(sel_features)}')
print(f'Features with coefficients shrunk to 0: {np.sum(lasso.estimator_.coef_==0)}')
```

```python
sel_features
```

```python
X_train=X_train[sel_features]
X_train
```
